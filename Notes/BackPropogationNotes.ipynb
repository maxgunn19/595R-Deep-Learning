# **Back Propogation (01/15/2026)**

**Algoritihimic Differentiation (AD)**  
  
*---forward mode---*  
scales with the number of inputs  

*---reverse mode---*  
scales with the number of outputs. one output only has to run once. perfect for **DL** (only 1 output, the *loss function*)  
  
**Backprop**  
*input:* x, y
*output:* a, b

*internals:* $a = x^2 y$ and $b = x \cos(y)$

*things with a bar:* $\bar{\zeta} = \frac{dL}{d\zeta}$  (derivative of loss with respect to the variable)  
  
In backprop $\bar{a}$ and $\bar{b}$ are known (just some number)  
  
With $\frac{dL}{da}$ and $\frac{dL}{db}$ being known how is $\bar{x}$ and $\bar{y}$ found?  
  
$$
\bar{x} = \frac{dL}{dx} = \frac{dL}{da}\frac{\delta a}{\delta x} + \frac{dL}{db}\frac{\delta b}{\delta x}
$$
$$
\bar{y} = \frac{dL}{dy} = \frac{dL}{da}\frac{\delta a}{\delta y} + \frac{dL}{db}\frac{\delta b}{\delta y}
$$
<center> *reverse form of the chain rule* 
</center>
  
**for $\mathbf{\bar{x}}$**  
$\frac{dL}{da}$ is $\qquad \mathbf{\bar{a}}$  
$\frac{\delta a}{\delta x}$ is $\qquad \mathbf{2xy}$  
$\frac{dL}{db}$ is $\qquad \mathbf{\bar{b}}$  
$\frac{\delta b}{\delta x}$ is $\qquad \mathbf{\cos(y)}$  
  
**for $\mathbf{\bar{y}}$**  
$\frac{dL}{da}$ is $\qquad \mathbf{\bar{a}}$  
$\frac{\delta a}{\delta y}$ is $\qquad \mathbf{x^2}$  
$\frac{dL}{db}$ is $\qquad \mathbf{\bar{b}}$  
$\frac{\delta b}{\delta y}$ is $\qquad \mathbf{-x\sin(y)}$
